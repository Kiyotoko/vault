## Josephus-Problem

- $n$ nummerierte Objekte werden im Kreis angeordnet.
- Jedes $k$-te Objekt wird entfernt (beginnend mit Objekt $k$)
- Nach jeder Runde wird der Kreis neu geschlossen
- Bei gegebenen $n,k\in\mathbb{N} (n\ge k)$ soll das letzte Objekt der Josephus-Permutation bestimmt werden

```python
def josephus_permutation(n: int, k: int) -> list[int]:
    l: list[int] = [i for i in range(1, n + 1)]
    r: list[int] = []
    i: int = 0
    while len(l) > 0:
        i+=1
        if (i < k):
            l.append(l[0])
        else:
            r.append(l[0])
            i = 0
        l.remove(l[0])
    return r

def josephus_last(n: int, k: int) -> int:
    if n == 1:
        return 1
    return ((josephus_last(n=n-1, k=k)+k-1) % n) + 1
```

```python
(min, max, k) = (1, 100, 3)
for n in range(min, max + 1):
    print(n, k, josephus_permutation(n, k))
```

```python
print(josephus_permutation(n=19, k=4))
print(josephus_last(n=19,k=4))
```

# Iris

```python
vocals = ['a', 'e', 'i', 'o', 'u', 'ö', 'ü', 'ä']
doubles = ['ei', 'au', 'eu']
```

```python
def löffel(i: str):
    s: str = ""
    for n in i.lower():
        s += n
        if n in vocals:
            s += 'lef' + n
    return s
```

```python
print(löffel("ich heise iris."))
```

```python
def ei(i: str):
    s: str = ""
    for n in i.lower():
        if n in vocals:
            s += 'ei'
        else:
            s += n
    return s
```

```python
print(ei("ich heise iris."))
```
